## SQL Analysis: 
This analysis explores the high risk regions, migration routes, most frequent incidents, and total deaths spanning 2014-2023.  

```python
import sqlite3
conn = sqlite3.connect("MissingMigrants.db")
```

```python
df.to_sql("MissingMigrants", conn, if_exists = "replace", index = False)
```

### Number of Dead and Missing based on the Region of Incident ###

```python
query = """
SELECT "region_of_incident", SUM("number_of_dead") as deaths, SUM(minimum_estimated_number_of_missing) AS est_missing
FROM MissingMigrants
GROUP BY region_of_incident
ORDER BY deaths DESC;

"""
result = pd.read_sql_query (query, conn)
result
```

The top 5 regions of incidents were: Mediterranean (9703), Northern Africa (5702), North America (3707), Southern Asia (2272), and Western Africa (2180). 

### Leading Cause Of Death

```python
query = """
SELECT cause_of_death, COUNT("cause_of_death") AS value_count, SUM("number_of_dead") AS deaths, SUM(minimum_estimated_number_of_missing) AS est_missing
FROM MissingMigrants
GROUP BY cause_of_death
ORDER BY value_count DESC;
"""
result = pd.read_sql_query (query, conn)
result
```

The leading cause of death is drowning followed my mixed or unknown causes of death.

```python
df['cause_of_death'].value_counts()
```

```python
query = """
SELECT region_of_incident, cause_of_death, COUNT(*) AS incidents, SUM("number_of_dead") AS deaths, SUM(minimum_estimated_number_of_missing) AS est_missing
FROM MissingMigrants
GROUP BY region_of_incident, cause_of_death
ORDER BY region_of_incident, deaths DESC;
"""
result = pd.read_sql_query (query, conn)
result
```

### High Risk Migration Routes

```python
query = """
SELECT
    migration_route,SUM(number_of_dead) AS deaths, SUM(minimum_estimated_number_of_missing) AS est_missing
FROM MissingMigrants
GROUP BY migration_route
ORDER BY deaths DESC;
"""
result = pd.read_sql_query (query, conn)
result
```

The migration route with the highest deaths are Unknown (7412 deaths) followed by Central Mediterranean (7114), Sahara Desert Crossing (5617 deaths), and US-Mexico Border Crossing (4512 deaths). The number of missing migrants however is more frequently documented in the Central Mediterranean route with 14,902 missing.

```python
query = """
SELECT region_of_incident, migration_route, COUNT(*) AS count, SUM("number_of_dead") AS deaths, SUM(minimum_estimated_number_of_missing) AS est_missing
FROM MissingMigrants
GROUP BY migration_route
ORDER BY region_of_incident, count DESC;
"""
result = pd.read_sql_query (query, conn)
result
```

### Fatalities over Time

```python
query = """
SELECT incident_year, SUM("number_of_dead") AS deaths, SUM(minimum_estimated_number_of_missing) AS est_missing
FROM MissingMigrants
GROUP BY incident_year
ORDER BY incident_year
"""
result = pd.read_sql_query (query, conn)
result
```

Deaths and number of missing fluctuates througout the years with minimum at 1725 deaths and 1409 missing and maximums at 4296 deaths and 2766 missing.
